```python
from IPython.display import Image
import random
import pandas as pd
import numpy as np

Image("task_1.png")
```

### Генерація фіч

```python
a0, a1, a2, a3 = (random.randrange(-10, 10) for i in range(4))

x1 = np.array([random.uniform(0, 10) for i in range(20)])
x2 = np.array([random.uniform(0, 20) for i in range(20)])
x3 = np.array([random.uniform(0, 15) for i in range(20)])
y = np.array([a0 + a1 * x1[i] + a2 * x2[i] + a3 * x3[i] for i in range(len(x1))])

df = pd.DataFrame({"x1" : x1, "x2" : x2, "x3" : x3, "y" : y})
```

## Перегляд вихідного датафрейму

```python
df
```

### Нормалізація і відбір еталонних значень 

```python
normalized_df = round((df-df.mean())/df.std(), 3)

etalon_x1 = df.iloc[np.where(abs(normalized_df["x1"]) == min(abs(normalized_df["x1"])))[0].tolist()[0], 0]
etalon_x2 = df.iloc[np.where(abs(normalized_df["x2"]) == min(abs(normalized_df["x2"])))[0].tolist()[0], 1]
etalon_x3 = df.iloc[np.where(abs(normalized_df["x3"]) == min(abs(normalized_df["x3"])))[0].tolist()[0], 2]

etalon_y = a0 + a1 * etalon_x1 + a2 * etalon_x2 + a3 * etalon_x3
```

```python
print("Еталонні значення(etalon_x1, etalon_x2, etalon_x3, etalon_y) = ", [round(i, 3) for i in (etalon_x1, etalon_x2, etalon_x3, etalon_y)])
```

### Точка плану, що задовольняє критерію вибору оптимальності

```python
Image("point.png")
```

```python
point = min((df.y - etalon_y) ** 2)
```

```python
print("Точка плану, що задовольняє критерію вибору оптимальності = " + str(point))
```
